## Einfaches Beispiel

```python
import numpy as np # Wir benutzen ein Numpy array fuer die Daten die wir 'plotten' moechten! Deswegen wird hier numpy importiert.

# Beispieldaten:
x = np.linspace(0,5,11) #Reminder: liefert uns gleichmaeßig verteilte Zahlen über ein bestimmtes Intervall. 11 Werte in (0,5)
y = x**2 #Reminder: quadriert die alle Werte in dem Numpy-Array x
```

Hinweis: Benutz gerne Sift+Tab, um euch die Dokumentationen fuer die einzelnen Funktionen, anzeigen zu lassen.

```python
plt.plot()
plt.plot(x, y, 'b') # 'b' steht fuer die Farbe blau
plt.xlabel('X Achse Titel')
plt.ylabel('Y Achse Titel')
plt.title('Titel')
plt.show()
```

## Matplotlib objektorientierte Methode
Das bedeutet, dass wir Figure-Objekte instanziieren und dann Methoden oder Attribute auf diesem Objekt aufrufen.

### Einfuehrung in die objektorientierte Methode
Diese Art und Weise ist uebersichtlicher, wenn mehrere Grafiken in einer Ausgabe ausgegeben werden sollen.

Zu Beginn erstellen wir eine Instanz `figure`. Dann können wir diesem Objekt Achsen hinzufuegen:

```python
# Erzeugen der Instanz figure
fig = plt.figure()

# Hinzufuegen von einer Menge an Achsen zu dem Objekt figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # links, unten, Breite, Hoehe (im Intervall von 0 und 1)
# hier am besten ein bisschen mit den Werten rumspielen, um ein Gefuehl zu bekommen!
# Bzw. bei den ersten beiden Werten aendert sich erst was, wenn mehrere Plots in der Ausgabe sind (siehe spaeter).

# Plot zu den Achsen hinzufuegen
axes.plot(x, y, 'b')
axes.set_xlabel('X Label') # Achtung: Hier .set_xlabel() und nicht wie zuvor nur .xlabel()
axes.set_ylabel('Y Label')
axes.set_title('Titel')
```

Code ist etwas komplizierter, aber der Vorteil ist, dass wir jetzt die volle Kontrolle darüber haben, wo die einzelnen Achsen der Plots platziert sind. Wir koennen dadurch einfach weitere Plots hinzufuegen:

```python
# Erzeugen der Instanz figure
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Haupt-Koordinatensystem (Haupt-Plot)
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inneres Koordinatensystem (kleiner Plot)

# Haupt-Koordinatensystem 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X Label Plot 1')
axes1.set_ylabel('Y Label Plot 1')
axes1.set_title('Plot 1 Titel')

# inneres Koordinatensystem (kleiner Plot) 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X Label Plot 2')
axes2.set_ylabel('Y Label Plot 2')
axes2.set_title('Plot 2 Titel');
```

## subplots()
Das Objekt plt.subplots() fungiert als automatischer Achsenmanager.

Use cases:

```python
# Aehnlich wie plt.figure(). Außer, dass hier ein Konzept aus Python zu Nutze genommen wird: 'tuple unpacking'
fig, axes = plt.subplots()
#'tuple unpacking' ermoeglicht es, einem Tupel von Variablennamen (hier: fig, axes), 
# Werte von einem Tupel, rechts neben der Zuweisung, zuzuweisen.

# Jetzt koennen wir dem axes Objekt Dinge (Plot, Label, Titel, etc.) hinzufuegen
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title') # mit ; keine Ausgabe von Text....
```

Eigentlicher Anwendungsfall: Wir koenen mit subplots() die Anzahl der Zeilen und Spalten den beiden Objekten uebergeben.

```python
# Ausgabefeld das zwei noch leere Sub-Plots nebeneinander hat
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# Hier in dem Fall ist axes nun ein Achsen-Array (noch leeres)
axes
```

```python
# Durch dieses Array koenen wir iterieren und dadurch Plot, Titel, Label hinzuefuegen
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Titel')

# Gebe das fig Objekt zurück    
fig
```

Ein haeufiges Problem bei Matplolib sind ueberlappende Sub-Plots oder Figuren. Durch die Methode `.tight_layout()` koennen wir das verhindern. Matplotlib ordnet dabei automatisch die Sub-Plots so an, dass sie nicht mehr ueberlappen:

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Titel')

fig    
fig.tight_layout() #plt.tight_layout() wuerde zum gleichen Ergebnis fuehren. Es ist daher egal was man verwendet
```

### Abbildungsgröße (figure size), Seitenverhältnis (aspect ratio) und DPI
Matplotlib ermoeglicht die Angabe von Seitenverhaeltnis, DPI und Abbildungsgroeße beim Erstellen des Figure-Objekts. 

   - `figsize` ist ein Tupel aus der Breite und Hoehe der Figur in Zoll.
   -  `dpi` ist die Punktzahl pro Zoll (Pixel pro Zoll).

Zum Beispiel:

```python
fig = plt.figure(figsize=(10,5), dpi=200)
```

Diese Parameter koenne auch dem Layout-Manager wie z.B: `subplots` uebergeben werden:

```python
fig, axes = plt.subplots(figsize=(10,5)) #gerne ein bisschen mit den Werten rumspielen

axes.plot(x, y, 'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Titel')
```

### Grafiken speichern
Matplotlib kann hochwertige Ausgaben in verschiedenen Formaten erzeugen, darunter PNG, JPG, EPS, SVG, PGF und PDF.

Um eine Grafik zu speichern kann man die  `savefig` Methode aufrufen:

```python
fig.savefig("filename.jpg")

# Hier kann DPI auch spezifiziert werden
fig.savefig("filenameDPI200.png", dpi=200)
```

## Legenden

Man kann den Parameter `label="label text"` verwenden, wenn Plots oder andere Objekte zur Figur hinzugefuegt werden. Wenn man das macht, kann man die Methode `legend()` ohne Parameter aufrufen, um eine Legende zur Figur hinzuzufuegen:

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")

ax.legend()
```

Falls die Legende ein Teil der Funktionen ueberlappen sollte, kann man die Position der Legende mit dem Paramter `loc` wie folgt beinflussen:

```python
ax.legend(loc=1) # oben rechts
ax.legend(loc=2) # oben links
ax.legend(loc=3) # unten links
ax.legend(loc=4) # unten rechts

# Es gibt viel mehr. Dokumentation: https://matplotlib.org/users/legend_guide.html#legend-location

# Am wichtigsten, ist folgende:
ax.legend(loc=0) # Hier entscheidet matplotlib automatisch, wo es am besten ist.
fig
```

## Einstellen von Farben, Linienbreiten und Linientypen

```python
# wie in MATLAB: Farbe und Linientyp
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blaue Linie mit Punkten
ax.plot(x, x**3, 'g--') # gruene gestrichelte Linie
```

### Color-Parameter

```python
fig, ax = plt.subplots()

ax.plot()
ax.plot(x, x+1, color="blue", alpha=0.2) # halbtransparent
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

### Linienbreite und Linientypen

Um die Linienbreite zu aendern kann man `linewidth` oder kurz `lw` verwenden. Der Linientyp kann durch `linestyle` oder kurz `ls` veraendert werden:

```python
# Hier ein paar Moeglichkeiten

fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# hier verschiedene Linientypen
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# Marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# Markergroeße
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

## Plotbereich

Wir koennen den Beriech der Achsen mit den `set_ylim` und `set_xlim` Methoden des Achsen-Objekts aendern. Oder man verwendet `axis('tight')` um automatisch "tightly fitted" Achsenbereiche zu bekommen:

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default Achsenbereich")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("enge Achse")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("definierter Achenbereich");
```
