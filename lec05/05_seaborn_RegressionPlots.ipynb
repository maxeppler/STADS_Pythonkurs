# Regressionsplots

Seaborn hat viele eingebaute Fähigkeiten für Regressionsplots. Hier werden wir uns lediglich die Funktion **lmplot()** anschauen und das Thema Regression im Allgemeinen nicht näher behandeln.

**lmplot** ermöglicht es, lineare Modelle anzuzeigen, aber es ermöglicht auch, diese Diagramme bequem nach Merkmalen aufzuteilen und den Farbton nach Merkmalen einzufärben.

Beispiele:

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## lmplot()

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

### Marker

lmplot kwargs werden an regplot übergeben, was eine umfassendere Form von lmplot() ist. regplot hat einen scatter_kws-Parameter, der an plt.scatter übergeben wird. Man setzt also den Parameter s in diesem Dictionary, der der quadratischen Markergröße entspricht. Mit anderen Worten, man übergibt ein Dictionary mit den Basisargumenten matplotlib, in diesem Fall s für die Größe eines Streudiagramms. Das muss man nicht im Kopf behalten, sondern kann stets in der Dokumentation nachschauen.

```python
# http://matplotlib.org/api/markers_api.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```

## Benutzung eines Gitters

Wir können eine variable Trennung durch Spalten und Zeilen eines Gitters hinzufügen. Dazu gibt man einfach die Argumente col oder row an.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```

## Aspekt und Größe

Seaborn-Figuren können ihre Größe und ihr Seitenverhältnis mit den Parametern **size** und **aspect** einstellen:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6,size=8)
```
