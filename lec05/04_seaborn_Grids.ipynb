# Grids (Gitter)

Gitter sind allgemeine Arten von Diagrammen, die es ermöglichen, Diagrammtypen auf Zeilen und Spalten eines Gitters abzubilden, was hilft, ähnliche Diagramme zu erstellen, die durch Merkmale (Features) getrennt sind.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## PairGrid

Pairgrid ist ein Teilplot-Gitter zum Darstellen von paarweisen Beziehungen in einem Datensatz.

```python
# Hier nur das Gitter noch ohne Daten
sns.PairGrid(iris)
```

```python
# Nun kann man die Daten zu dem Gitter hinzufuegen
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# Einstellen welche Darstellung zu upper,lower und diagonal
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

## pairplot

pairplot ist eine einfachere Version von PairGrid (wird sehr haeufig verwendet).

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
```

## Facet Grid

FacetGrid wird dazu verwendet, um Raster von Plots zu erstellen, die auf einem Feature basieren:

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Nur das Gitter bzw Raster
g = sns.FacetGrid(tips, col="time", row="smoker")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
# Siehe hier die Anordnung der Argumente
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

## JointGrid

JointGrid ist eine weitere Möglichkeit Plots vom Typ jointplot() zu erstellen. Kurzes Beispiel:

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```
